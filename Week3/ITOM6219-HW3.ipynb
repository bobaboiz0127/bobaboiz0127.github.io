## Step 8. Topic Discovery
We use non-negative matrix factorization to generate 20 topics.
For each topic, we use the top 3 words to describe the topic.

````python
# Apply NMF
from sklearn.decomposition import NMF
# 💡 Hint: See class notebook Section 3.4 "Topic Modeling" – use n_components=20 here (class used 7)
#         The for-loop to print topic names is also shown there

nmf_openAI = NMF(
    n_components=10,
    random_state=0,
    init='nndsvda',
    max_iter=1000
)
W_openAI = nmf_openAI.fit_transform(tfidf_openAI)

# ---------- Create topic names ----------
feature_names_openAI = tfidf_vectorizer_openAI.get_feature_names_out()

topic_names_openAI = []
for topic_idx, topic in enumerate(nmf_openAI.components_):
    top_words = [feature_names_openAI[i] for i in topic.argsort()[:-4:-1]]
    topic_label = f"Topic_{topic_idx+1}_" + "_".join(top_words)
    topic_names_openAI.append(topic_label)
    print(f"{topic_label}: {', '.join(top_words)}")

# ---------- Topic dataframe ----------
topic_df_openAI = pd.DataFrame(W_openAI, columns=topic_names_openAI)
topic_df_openAI.head()
````

````python
# Step 10: Repeat topic modeling and regression for OpenAI impressions

def run_topic_regression(df, outcome_col, n_components=20, max_df=0.8, min_df=0.02):
    docs = df['text'].fillna('')
    vectorizer = TfidfVectorizer(stop_words='english', max_df=max_df, min_df=min_df)
    tfidf_matrix = vectorizer.fit_transform(docs)

    nmf = NMF(n_components=n_components, random_state=0, init='nndsvda', max_iter=1000)
    W_local = nmf.fit_transform(tfidf_matrix)

    feature_names_local = vectorizer.get_feature_names_out()
    local_topic_names = []
    for topic in nmf.components_:
        top_words = [feature_names_local[i] for i in topic.argsort()[:-4:-1]]
        local_topic_names.append("_".join(top_words))

    X_local = pd.DataFrame(W_local, columns=local_topic_names)
    y_local = np.log1p(df[outcome_col])

    regression_results = pg.linear_regression(X_local, y_local)
    return regression_results.round(3), local_topic_names

results_impression_openAI, topic_names_impression_openAI = run_topic_regression(
    tweets_openAI,
    'public_metrics.impression_count',
    n_components=20
)

print("OpenAI regression results for log(Impressions + 1):")
print(results_impression_openAI)
````
